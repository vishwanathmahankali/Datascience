```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,11)
x
```

```python
y = 2 *x
y
```

```python
y1 = 2 * x
y2 = 3 * x
y1,y2
```

```python
plt.plot(x,y1, color = 'green', linewidth = 2, linestyle = '--')
plt.plot(x,y2, color = 'red', linewidth = 2, linestyle = '--')
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
student = {"Sam": 30, "Bob": 50, "Julia": 70}
names = list(student.keys())
marks = list(student.values())
names, marks
```

```python
plt.bar(names, marks, color = "yellow")
plt.title("Marks of Students")
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.grid(True)
plt.show()
```

```python
plt.barh(names, marks, color = "yellow")
plt.title("Marks of Students")
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.grid(True)
plt.show()
```
